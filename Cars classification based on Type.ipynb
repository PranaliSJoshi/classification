# Divide Categorical and Continuous columns

```python
cat = []
con = []
for i in A.columns:
    if(A[i].dtypes =="object"):
        cat.append(i)
    else:
        con.append(i)
```

# Standardization using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit_transform(A[con])
```

```python
pd.DataFrame(ss.fit_transform(A[con]))
```

```python
B = pd.DataFrame(ss.fit_transform(A[con]),columns=con)
```

# Standardization using MinMaxScaler
